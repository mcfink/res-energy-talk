```python
import numpy as np
import pandas as pd
```

```python
fuel_mix_2017 = pd.read_csv(f'../raw_data_files/fuel_mix/2017_final_fuel_mix.csv')
fuel_mix_2018 = pd.read_csv(f'../raw_data_files/fuel_mix/2018_final_fuel_mix.csv')
marginal_fuels_2017 = pd.read_csv(f'../raw_data_files/fuel_mix/2017_marginal_fuels.csv')
marginal_fuels_2018 = pd.read_csv(f'../raw_data_files/fuel_mix/2018_marginal_fuels.csv')
```

```python
fuel_mix_df = fuel_mix_2017.append(fuel_mix_2018)
del fuel_mix_df['Marginal']
marginal_fuel_df = marginal_fuels_2017.append(marginal_fuels_2018)
all_df = fuel_mix_df.merge(marginal_fuel_df, on='Begin Date')
```

```python
all_df.to_csv('fuel_mix_w_marginal.csv')
```
