```python
import numpy as np
import pandas as pd
import pytz
```

```python
a = pd.read_csv('raw_data_files/insolation_data/ESXV1_mod_2019-01-01.csv')
```

```python
a['pandas_dt'] = a['Date_Time'].apply(lambda x: pd.to_datetimeatetime(x))
```

```python
a['local_time'] = a['pandas_dt'].apply(lambda x: pytz.utc.localize(x).astimezone(pytz.timezone('US/Eastern')))
```

```python
a = a.set_index('local_time')
```

```python
one_minute = a.resample('1T').interpolate('linear')
one_minute.head(20)
```

```python
fifteen_minute_df = one_minute.resample('15T').asfreq()
fifteen_minute_df[['solar_radiation_Wperm2']].reset_index().to_csv('insolation.csv')
```
